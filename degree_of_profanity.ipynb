```python
# importing required libraries
import pandas as pd
```

```python
# Storing tweets in dataframe and profane words in list
profane_words = ['Sucks', 'Hate']
tweets = pd.DataFrame({'tweets' : ['Monday SUCKS!!!!', 'I hate sucking mondays', 'Have a good day every body:)']})
```

```python
tweets
```

```python
# converting all letter to small case
profane_words = list(map(lambda x: x.lower(), profane_words))
tweets['tweets'] = tweets['tweets'].apply(lambda x: x.lower())
```

```python
tweets
```

```python
# Split the tweets to lists of words. This will help us to do futher preprocessing
tweets['tweets'] = tweets['tweets'].str.split()
```

```python
tweets
```

```python
#we will use stemming to reduce noise in words

from nltk.stem.snowball import SnowballStemmer

# Use English stemmer.
stemmer = SnowballStemmer("english")

tweets['tweets_stemmed'] = tweets['tweets'].apply(lambda x: [stemmer.stem(word) for word in x])
```

```python
tweets
```

```python
# degree of profanity can be said total profane words appear in sentence wrt to how much value they hold. comapring it with total words after removing the stopwords(a, an, the, I, etc) and punctuation
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
all_stopwords = stopwords.words('english')

tweets['tweets_stemmed_'] = tweets['tweets_stemmed'].apply(lambda x: [word for word in x if not word in all_stopwords])
```

```python
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r'\w+')
tweets['tweets_stemmed_tokenised'] = tweets['tweets_stemmed_'].apply(lambda x: " ".join(x))
tweets['tweets_stemmed_tokenised'] = tweets['tweets_stemmed_tokenised'].apply(lambda x: tokenizer.tokenize(x))
```

```python
tweets
```

```python
def degree_of_profanity(word_list):  #ratio of total profane words appeared in sentence and and total useful words
  pw = sum(ele in word_list for ele in profane_words)
  return pw/len(word_list)


tweets['degree_of_profanity'] = tweets['tweets_stemmed_tokenised'].apply(degree_of_profanity)
```
